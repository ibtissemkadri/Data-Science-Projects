```python
import re
```

```python
def quant(x):
    quant=re.findall(r'[\d,.]+',x)
    
    if len(quant)>=1:
        if quant[0].find(',')!=-1:
            quant[0]=quant[0].replace(',','.')
        return float(quant[0])
    else:
        return 0
```

```python
def gram(x):
    x=x.lower()
    unit_l=re.findall(r'[\w\s,.]+g',x)
    if len(unit_l)>=1:
        return unit_l[0]
    else:
        return np.nan
```
