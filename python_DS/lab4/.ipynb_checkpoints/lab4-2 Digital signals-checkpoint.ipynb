## 2/ Dessiner le Signal et son spectrogramme

```python
data=wav_file[1]
```

```python
plt.plot(data)
```

```python
plt.plot(data[:1000])
```

```python
from scipy.fftpack import fft, ifft, fftshift
```

```python
nfft=512
fft_sig = fft(data,nfft)
f=np.arange(0,fs,fs/nfft)
spec=np.abs(fft_sig)
plt.plot(f,spec)
```

```python
f=np.arange(0,fs/2,fs/nfft)
spec=np.abs(fft_sig)
plt.plot(f,spec[:nfft//2])
```

```python
from scipy import signal
```

```python
fr, t, Sxx = signal.spectrogram(data, fs)
plt.pcolormesh(t, fr, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```

```python
Energy=np.power(spec,2)
plt.plot(f,Energy[:nfft//2])
```

## 3/ La fenêtre réctangulaire

```python
a=3000
b=8000
rect=np.array([1 if (x>=a and x<=b) else 0 for x in range(data.shape[0])])
plt.plot(rect[a//2:b*2])
```

## 4/ Troncature du signal

```python
rect_signal=data*rect
plt.plot(rect_signal[a//2:b*2])
```

```python
fft_rect_sig=fftshift(fft(rect_signal))
spec_rect_sig=np.abs(fft_rect_sig)
plt.plot(spec_rect_sig)
```

## 5/ Tranformée de fourrier de la fenêtre

```python
fft_rect = fftshift(fft(rect))
plt.plot(np.abs(fft_rect[a:b])[:500])
```

```python
conv_rect = signal.convolve(fft_sig, fft_rect)
plt.plot(np.abs(conv_rect))
```

## 6/ Filtres

```python
filt_med= signal.medfilt(data)
filtered_med=np.abs(ifft(filt_med))
```

```python
wavfile.write("medfilt.wav", fs, filtered_med)
```

```python
filt_hilbert = signal.hilbert(data)
filtered_hilbert=np.abs(ifft(filt_hilbert))
```

```python
wavfile.write("hilbertfilt.wav", fs, filtered_hilbert)
```
