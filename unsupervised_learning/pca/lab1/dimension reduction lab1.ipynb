```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
datetime.strptime('2015-08-14 17',"%Y-%m-%d %H").timestamp()
```

```python
df['timestamp']=pd.to_datetime(df['date'], infer_datetime_format=True).values.astype(np.int64) // 10 ** 9
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df)
```

```python
pca=PCA(n_components=len(df.columns))
pca_model=pca.fit_transform(scaled_df)
```

```python
exp_var=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4))
var=pca.explained_variance_ratio_
plt.step(range(1, len(df.columns)+1), exp_var, color='k')
plt.bar(range(1, len(df.columns)+1),var, color='r')
```
