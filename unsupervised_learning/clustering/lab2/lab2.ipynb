```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Generating data

```python
from sklearn.datasets import make_circles, make_blobs, make_moons
```

```python
X,y=make_circles(n_samples=500)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

# Clustering

```python
from sklearn.cluster import MiniBatchKMeans, MeanShift, SpectralClustering, estimate_bandwidth
from sklearn.mixture import GaussianMixture
```

```python
miniBatch=MiniBatchKMeans(n_clusters=2)
miniBatch.fit(X_scaled)
plt.scatter(X_scaled[:,0],X_scaled[:,1], c=miniBatch.labels_)
```

```python
bandwidth = estimate_bandwidth(X, quantile=0.5, n_samples=300)
meanShift=MeanShift(bandwidth = bandwidth, bin_seeding=True)
meanShift.fit(X_scaled)
plt.scatter(X_scaled[:,0],X_scaled[:,1], c=meanShift.labels_)
```

```python
spect=SpectralClustering(n_clusters=2, affinity='nearest_neighbors')
spect.fit(X_scaled)
plt.scatter(X_scaled[:,0],X_scaled[:,1], c=spect.labels_)
```

```python
gauss=GaussianMixture(n_components=2)
gauss.fit(X_scaled)
labels=gauss.predict(X_scaled)
plt.scatter(X_scaled[:,0],X_scaled[:,1], c=labels)
```
