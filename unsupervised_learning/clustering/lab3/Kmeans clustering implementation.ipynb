```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Distance computation

```python
def dist(x,y):
    return np.linalg.norm(x-y, axis=1)
```

# Means of clusters

```python
def bary(pts):
    m=pts.shape[0]
    return pts.sum(axis=0)/m
```

# Centroids initialization

```python
def randp(pts, k):
    ind= np.random.choice(pts.shape[0], k, replace=False)
    return pts[ind,:]
```

# K means implementation

```python
def km(X, K):
    m=X.shape[0]
    centroids=randp(X, K)
    arr=10
    while(arr>10**(-9)):
        labels=np.empty((m))
        for i in range (m):
            labels[i]=np.argmin(dist(X[i,:], centroids))
        old_cent=centroids.copy()
        for k in range(K):
            centroids[k,:]=bary(X[labels==k,:])
        arr=np.linalg.norm(centroids-old_cent)
    return centroids, labels
```

# Application

## Generating data

```python
from sklearn.datasets import make_blobs
X,y= make_blobs(n_samples=300, n_features=3, centers=4, cluster_std=1.4, random_state=101)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

## Our algorithm

```python
centers, labels=km(X_scaled,4)
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels)
```

## With Sklearn

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4)
kmeans.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=kmeans.labels_)
```

## Comparing results

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(15,7))
ax1.set_title('implemented K_means')
ax1.scatter(X_scaled[:,0],X_scaled[:,1],c=labels,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X_scaled[:,0],X_scaled[:,1],c=y,cmap='rainbow')
ax3.set_title('K Means (sklearn)')
ax3.scatter(X_scaled[:,0],X_scaled[:,1],c=kmeans.labels_,cmap='rainbow')
```
