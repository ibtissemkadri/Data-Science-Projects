```python
import requests
import pandas as pd
import time
from datetime import datetime
from elasticsearch import Elasticsearch
```

```python
urls = ["https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2016,2015,2014&r=788&p=all&rg=2&cc=20&uitoken=dbd2afeda503239ad1a19b97449e4f99",
        "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2013,2012&r=788&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
        "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2016,2015&r=251&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
        "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2014,2013&r=251&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
       "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2012&r=251&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
       "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2016,2015&r=818&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
        "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2014,2013&r=818&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719",
       "https://comtrade.un.org/api/get?max=500&type=C&freq=A&px=HS&ps=2012&r=818&p=all&rg=2&cc=20&uitoken=4c599ed9273b53b875f870451acdf719"]
```

```python

years = []
trades = []
exp_country = []
destinations =[]
products = []
values = []
retailers='comtrade'
elastic_connection = Elasticsearch(['http://localhost:9200'])

for url in urls:
    req = requests.get(url)
    data = req.json()
    
    i = 0
    while (i < len(data['dataset'])):
        item = {}
        years.append(data['dataset'][i]['yr'])
        trades.append(data['dataset'][i]['rgDesc'])
        exp_country.append(data['dataset'][i]['rtTitle'])
        destinations.append(data['dataset'][i]['ptTitle'])
        products.append(data['dataset'][i]['cmdDescE'])
        values.append(data['dataset'][i]['TradeValue'])
        
        item['partners'] = data['dataset'][i]['ptTitle']
        item['reporters'] = data['dataset'][i]['rtTitle']
        item['product'] = data['dataset'][i]['cmdDescE']
        #item['trade'] = data['dataset'][i]['rgDesc']
        item['trade_values'] = int(data['dataset'][i]['TradeValue'])
        item['years'] = data['dataset'][i]['yr']
        item['retailers']=retailers
        dt = time.time()
        doc_id = "%s-%s-%s"%('comtrade.org',dt,i)
        elastic_connection.index(index='trademap',doc_type='export',id= doc_id, body=item)
                    
        i+=1
        
    time.sleep(1)
index = [ j for j in range(len(years))]  
```

```python
df = pd.DataFrame({'years':years,'reporters':exp_country, 'partners':destinations, 'products':products, 'trade_values':values}, index)
```
